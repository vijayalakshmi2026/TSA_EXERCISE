```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic time-series data
np.random.seed(42)
days = 365  # Number of days
date_range = pd.date_range(start='2023-01-01', periods=days, freq='D')

# Generate temperature data with seasonal trends and noise
temperature = 25 + 5 * np.sin(np.linspace(0, 2 * np.pi, days)) + np.random.normal(0, 1, days)

# Create DataFrame
data = pd.DataFrame({'Date': date_range, 'Temperature': temperature})
data.set_index('Date', inplace=True)

# Apply Moving Average Smoothing
window_size = 7  # Number of days for smoothing
data['Smoothed_Temperature'] = data['Temperature'].rolling(window=window_size).mean()

# Plot Original vs Smoothed Data
plt.figure(figsize=(15, 5))
plt.plot(data.index, data['Temperature'], label="Original Temperature", color="green", alpha=0.5)
plt.plot(data.index, data['Smoothed_Temperature'], label=f"{window_size}-Day Moving Average", color="black", linewidth=2)
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Moving Average Smoothing for Time Series Data")
plt.legend()
plt.show()
```
