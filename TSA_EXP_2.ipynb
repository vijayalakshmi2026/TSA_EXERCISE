```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate synthetic time series data
np.random.seed(0)
dates = pd.date_range(start='2020-01-01', periods=365, freq='D')  # 365 days, daily frequency
data = np.random.randn(365) * 10 + 100  # Random data around 100 with some noise

# Create a DataFrame
df = pd.DataFrame(data, columns=['Value'], index=dates)

# Display first few rows
df.head()
```

```python
# Plot original time series data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Value'], label='Original Data', color='blue')
plt.title('Original Time Series Data', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Create a 7-day moving average
df['7_day_MA'] = df['Value'].rolling(window=7).mean()

# Plot original data and moving average
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Value'], label='Original Data', color='blue', alpha=0.5)
plt.plot(df.index, df['7_day_MA'], label='7-Day Moving Average', color='orange', linewidth=2)
plt.title('Time Series with 7-Day Moving Average', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Resample the data to monthly frequency using the mean
df_monthly = df.resample('M').mean()

# Plot resampled data
plt.figure(figsize=(12, 6))
plt.plot(df_monthly.index, df_monthly['Value'], label='Monthly Average', color='green', linewidth=2)
plt.title('Monthly Resampled Time Series Data', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot histogram of the original data
plt.figure(figsize=(12, 6))
sns.histplot(df['Value'], bins=30, kde=True, color='purple')
plt.title('Distribution of Time Series Data', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Box plot of the original data
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Value'], color='cyan')
plt.title('Box Plot of Time Series Data', fontsize=14)
plt.xlabel('Value')
plt.grid(True)
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot

# Autocorrelation plot
plt.figure(figsize=(12, 6))
autocorrelation_plot(df['Value'])
plt.title('Autocorrelation Plot', fontsize=14)
plt.show()
```
