```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dates=pd.date_range(start="2025-01-01",end="2025-01-10",freq="D")
```

```python
data={
    'Date':dates,
    'Temperature':[22.1, 23.4, np.nan, 25.6, np.nan, 28.3, 29.4, np.nan, 31.0, 32.5],
    'Humidity':[60, 62, 65, np.nan, 68, np.nan, 70, 72, np.nan, 75]
}
df=pd.DataFrame(data)
```

```python
df.set_index('Date',inplace=True)
```

```python
print(df)
```

```python
df_interpolated = df.interpolate(method='linear')
plt.figure(figsize=(12, 6))
```

```python
plt.plot(df.index, df['Temperature'], label='Original Temperature', marker='o', color='red', linestyle='--', alpha=0.6, markersize=8)
plt.plot(df.index, df['Humidity'], label='Original Humidity', marker='o', color='blue', linestyle='--', alpha=0.6, markersize=8)
plt.title('Original Time Series Data with Missing Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df_interpolated['Temperature'], label='Interpolated Temperature', marker='o', color='green', linestyle='-', markersize=6)
plt.plot(df.index, df_interpolated['Humidity'], label='Interpolated Humidity', marker='o', color='orange', linestyle='-', markersize=6)
plt.title('Interpolated Time Series Data (After Interpolation)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
